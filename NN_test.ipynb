```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np
import pandas as pd

from NN import FullyConnectedNet
from solver import Solver

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
data = {'X_train':X_train, 'y_train':y_train}
f' dtype = {type(X_train)}, x_shape = {X_train.shape}, y_shape = {y_train.shape}'
```

```python
model = FullyConnectedNet(hidden_dims=[8], input_dim=X.shape[1], reg=0, num_classes=3)
FCNet = Solver(model, data, learning_rate=1e-2, num_epochs=20)
FCNet.train()
y_pred = np.argmax(model.loss(X_test), axis=1)
np.mean(y_pred == y_test) * 100
```

```python
import matplotlib.pyplot as plt

plt.plot(FCNet.loss_history)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss History")
plt.figure(figsize=(10,6))
```

```python
plt.figure(figsize=(10,6))
plt.plot(FCNet.train_acc_history)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```
